### Comparing Optimal Bayes and Naive Bayes using simulated Gaussian data for non-zero covariance.

#### Imports

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
import seaborn as sns
import itertools
from scipy.stats import norm
from scipy.stats import multivariate_normal
import scipy.stats
from sklearn.naive_bayes import GaussianNB

%matplotlib inline
sns.set()
```

#### Data Generation

```python
# # Data generation

# #data of Gaussian Distributions
# #distribution 1
# mu1 = [2, 2]
# sigma1 = [[5, 0], [0, 5]] # Notice the 0 covariance
# x1, y1 = np.random.multivariate_normal(mu1, sigma1, 100).T

# #distribution 2
# mu2 = [-2, -2]
# sigma2 = [[3, 0], [0, 3]] # Notice the 0 covariance
# x2, y2 = np.random.multivariate_normal(mu2, sigma2, 100).T

#data of Gaussian Distributions
np.random.seed(seed = 3)
#distribution1
mu1 = [2, 2]
sigma1 = [[5, -4], [-4, 5]]
x1, y1 = np.random.multivariate_normal(mu1, sigma1, 100).T
#distribution2
mu2 = [-2, -2]
sigma2 = [[3.5, 3], [3, 3.5]]
x2, y2 = np.random.multivariate_normal(mu2, sigma2, 100).T
```

#### Excercise: A) Plot the decision boundary

```python
#%%%%%%%%%%%%%% Excercise: Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the decision boundary for the optimal and the naive Bayes classifiers goes here
"""


# 2D data distribution
N = 100
X = np.linspace(-8, 8, N)
Y = np.linspace(-8, 8, N)
X, Y = np.meshgrid(X, Y)

# setup of figure
fig = plt.figure(figsize=(20, 10))

#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%#
ax = fig.add_subplot(1,2,1)

# plotting contours of the gaussians on the left plot
zz = multivariate_normal.pdf(np.vstack([X.ravel(), Y.ravel()]).T, mean=mu1, cov=sigma1)
Z = zz.reshape(X.shape)
ax.contour(X, Y, Z, 5, alpha=0.3, colors = 'darkblue')

zz = multivariate_normal.pdf(np.vstack([X.ravel(), Y.ravel()]).T, mean=mu2, cov=sigma2)
Z = zz.reshape(X.shape)
ax.contour(X, Y, Z, 5, alpha=0.3, colors = 'darkgreen')


ax.scatter(x1, y1, alpha=0.7, color='blue')
ax.scatter(x2, y2, alpha=0.7, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Scatter plot of two Gaussians')

#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%#

# right side plot - Optimal Bayes Classifier
ax = fig.add_subplot(1,2,2)

zz1 = multivariate_normal.pdf(np.vstack([X.ravel(), Y.ravel()]).T, mean=mu1, cov=sigma1)
zz2 = multivariate_normal.pdf(np.vstack([X.ravel(), Y.ravel()]).T, mean=mu2, cov=sigma2)
Z = np.argmax(np.array([zz1, zz2]), axis=0).reshape(X.shape)

# plot the contours
ax.contour(X, Y, Z, 1, alpha=0.3, colors = ('darkblue','darkgreen'))
ax.contourf(X, Y, Z, 1, alpha=0.3, colors = ('darkblue','darkgreen'))

# x, y given coordinates concatenated to X_data
X_data  = np.vstack([np.vstack([x1, y1]).T, np.vstack([x2, y2]).T])
# Y_data is the class of data
Y_data = np.hstack([[0]*x1.shape[0], [1]*x2.shape[0]])

# fit model
model = GaussianNB()
model.fit(X_data, Y_data)

# prediction for each point on meshgrid
zz = model.predict(np.vstack([X.ravel(), Y.ravel()]).T)
Z = zz.reshape(X.shape)

# plot the contours
ax.contour(X, Y, Z, 2, alpha=0.3, colors='red')

ax.scatter(x1, y1, alpha=0.7, color='blue')
ax.scatter(x2, y2, alpha=0.7, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Optimal (blue) vs Naive (red) Bayes')

plt.show()
```

#### Excercise: B) Accuracies

```python
#%%%%%%%%%%%%%% EXCERCISE - Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the accuracies of the optimal and the naive Bayes classifiers goes here
"""
#Optimal Bayes accuracy
y_pred1 = multivariate_normal.pdf(X_data, mean=mu1, cov=sigma1)
y_pred2 = multivariate_normal.pdf(X_data, mean=mu2, cov=sigma2)
y_pred = np.argmax(np.vstack([y_pred1, y_pred2]), axis=0)
print('Optimal Bayes Accuracy:', np.mean(y_pred == Y_data))

#Naive Bayes accuracy
print('Naive Bayes Accuracy:', model.score(X_data, Y_data))
```
