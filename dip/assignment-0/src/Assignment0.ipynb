# Learning about Jupyter Notebooks and basic libraries

## Loading libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_folder = './../images'
```

## Loading images using OpenCV

```python
# converts image from BGR to RGB
def RGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline

img1 = RGB(cv2.imread(image_folder + '/lena.png'))
img2 = RGB(cv2.imread(image_folder + '/lotus.jpg'))
img3 = RGB(cv2.imread(image_folder + '/gamma.jpg'))
img4 = RGB(cv2.imread(image_folder + '/bell.jpg'))

plt.imshow(img1)
```

## Display images using matplotlib

```python
fig = plt.figure(figsize=(8,8))

plt.subplot(2, 2, 1)
plt.imshow(img1)

plt.subplot(2, 2, 2)
plt.imshow(img2)

plt.subplot(2, 2, 3)
plt.imshow(img3)

plt.subplot(2, 2, 4)
plt.imshow(img4)

plt.show()
```

## Image to grayscale and saving image

```python
img = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
cv2.imshow('img', img)
# press any key to exit the image screen
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
# saving image
cv2.imwrite(image_folder + '/gray.png', img)
```

## Checking range of values

*This image has min value 3 and max value 255. This implies that each pixel in the image is represented by 8 bits or values from 0 to 255.*

```python
plt.hist(img1.ravel())
plt.show()
print('min value: ' + str(np.min(img1)))
print('max value: ' + str(np.max(img1)))
```
